```python
import pandas as pd
```

```python
book_metadata = pd.read_parquet('metadata_en.parquet')
book_metadata
```

```python
profanity_words = [
    'arsehole', 'asshat', 'asshole', 'bastard', 'big black cock',
    'bitch', 'bloody', 'blowjob', 'bollocks', 'bugger', 'bullshit',
    'chicken shit', 'clusterfuck', 'cock', 'cocksucker', 'coonass',
    'cornhole', 'cox-zucker machine', 'cracker', 'cunt', 'damn', 'dick',
    'enshittification', 'faggot', 'feck', 'fuck', 'fuckery', 'grab \'em by the pussy'
    'healslut', 'kike', 'motherfucker', 'nigga', 'nigger', 'paki', 'poof',
    'poofter', 'prick', 'pussy', 'ratfucking', 'retard', 'shit', 'shithouse',
    'shitposting', 'shitter', 'shut the hell up', 'slut', 'son of a bitch',
    'spic', 'taking the piss', 'twat', 'wanker', 'whore'
]
false_terms = [
    'dicken', 'sweatwater', 'sweetwater', 'spice'
]
title_subject = book_metadata[['Index', 'title', 'subjects']]
```

```python
def contains_profanity(text, profane_words):
    for word in profane_words:
        if word in text.lower():
            for false_word in false_terms:
                if word in false_word and false_word in text.lower():
                    return False

    return any(word in text.lower() for word in profane_words)
```

```python
title_subject.loc[:, 'contains_profanity'] = title_subject.apply(
    lambda row: contains_profanity(str(row['title']), profanity_words) or contains_profanity(str(row['subjects']), profanity_words),
    axis = 1
)
```

```python
profanity_books = title_subject[title_subject['contains_profanity']]
profanity_books
```
